# Will/Dinda write up

There are two main issues we need to handle:
1. Our goal is to predict an artist's success before they appear on one of the key playlists. However, our data also contains all the streams after the artist became successful. Hence, we need to drop post-success rows.
2. To achieve a good dataset for training, we have to ensure we'll have the same streaming timespan for each artist in the dataset. Thus, we should only pick one year for prediction. We will predict only successful artist in 2017.

So we wrote `getPriorData(data)` and `splitSucc2017(df)` functions that will handle those two issues.

```python

def getSuccessYear(data):
    # remove the particular stream of the artists if date < date streamed in key playlist
    nX = data[data['playlist_id'].isin(success_code)]
    data = data[~data['playlist_id'].isin(success_code)]
    a = nX.groupby('artist_name').agg({'year' : [np.min]})
    a = pd.DataFrame(a)
    data=data.merge(a, left_on='artist_name', right_on=a.index.values, how='left')
    data=data.rename(index=str, columns={('year', 'amin'): "success_year"})
    return data

def getSuccessDate(data):
    # get the date artist first streamed in key playlists
    nX = data[data['playlist_id'].isin(success_code)]
    data = data[~data['playlist_id'].isin(success_code)]
    a = nX.groupby('artist_name').agg({'date' : [np.min]})
    a = pd.DataFrame(a)
    data=data.merge(a, left_on='artist_name', right_on=a.index.values, how='left')
    data=data.rename(index=str, columns={('date', 'amin'): "success_date"})
    return data

def dropAfterSuccess(df):
    # drop data after the artist became successful
    df = df.drop(df[df['date'] > df['success_date']].index)
    #double protection: drop rows that contains key playlists
    df = df[df.playlist_id.isin(success_code) == False]
    
    return df

def splitSucc2017(df):
    print('dropping values from main dataframe')
    df = df.drop(df[df.success_year<2017.0].index)
    df = df.drop(['success_year'],axis=1)
    return df

# aggregate
def getPriorData(data):
    # get data prior to the artist's appearance on key playlist_id
    print('\nlocating prior success data')
    print('total full rows: '+ str(len(data)))
    data = getSuccessDate(data)
    data = dropAfterSuccess(data)
    data = data.drop(['success_date'],axis=1)
    print('total rows now: '+str(len(data)))
    return data
```

```python
#Stream count per artist
def calculateStreamCount(data):
    streams = data['artist_name'].groupby(data['artist_name']).count().to_dict()
    data['stream_count']=data['artist_name'].map(streams)
    return data

# Number of users per artist
def calculateUsersPerArtist(data):
    streams = data.groupby('artist_name')['customer_id'].nunique().to_dict()
    data['nunique_customer']=data['artist_name'].map(streams)
    return data

def calculatePassionScore(data):
    data['passion_score'] = data['stream_count']/data['nunique_customer']  
    return data

def getArtistFeatures(data):
    data = calculateStreamCount(data)
    data = calculateUsersPerArtist(data)
    data = calculatePassionScore(data)
    return data
```

## Playlist Features

Understanding an artist’s growth as a function of his/her movement across different playlists is potentially key to understanding how to identify and breakout new artists on Spotify. 

In turn, this could help us identify the most influential playlists and the reasons for their influence.

One way to model the effect of playlists on an artist’s performance has been to include them as categorical features in our model, to note if there are any particular playlists or combinations of playlists that are responsible for propelling an artist to future success:

### *Artist Feature 1 + Artist Feature 2 …. + Artist Feature N = Probability of Success*
**
Success (1) = Artist Features on Key Playlist
Failure (0) = Artist Not Featured on Key Playlist
**

Where,

**
⇒Artist Feature N = Prior Playlist 1 + Prior Playlist 2 +…Prior Playlist N
**


Given that we have over 19,000 playlists in our dataset or 600 artists, using the playlists each artist has featured on, as categorical variables would lead to too many features and a very large, sparse matrix. 

Instead, we need to think of ways to summarize the impact of these playlists. One way to do this would be to consider the top 20 playlists each artist has featured on. 

Even better would be to come up with one metric that captures the net effect of all top 20 prior playlists, for each artist, rather including using all 20 playlists for each artists as binary variables. The intuition here is that if this metric as a whole has an influence on the performance of an artist, it would suggest that rather than the individual playlists themselves, it is a combination of their generalized features that affects the future performance of an artist. 

Accordingly, different combinations of playlists could equate to having the same impact on an artist, thereby allowing us to identify undervalued playlists.

Some of the features such a metric could use is the number of unique users or ‘reach’, number of stream counts, and the passion score of each playlist

* Prior Playlist Stream Counts
* Prior Playlist Unique Users (Reach)
* Prior Playlist Passion Score

There are several other such features that you could generate to better capture the general characteristics of playlists, such as the average lift in stream counts and users they generate for artists that have featured on them.

# Write-up

Our plan is to prepare artist rows data as our dataset. Currently, our data is still stream-based, so if we just calculate the playlist statistics, we will lose some data once we dropped the duplicate artist names. But we can still calculate the playlist statistics per artist and add it as playlist features for each artist. This can be done through `getArtistPlaylistFeatures(data)`.

```python
def playlist_avg_stream_counts(data):
    streams = data['playlist_id'].groupby(data['playlist_id']).count().to_dict()
    data['playlist_stream_count']=data['playlist_id'].map(streams)
    data['playlist_stream_count']=data['playlist_stream_count'].fillna(0)
    return data

def playlist_avg_number_of_users(data):
    users = data.groupby('playlist_id')['customer_id'].nunique().to_dict()
    data['playlist_customers_no']=data['playlist_id'].map(users)
    data['playlist_customers_no']=data['playlist_customers_no'].fillna(0)
    return data
 
def playlist_avg_passion_score(data):
    data['playlist_passion_score'] = data['playlist_stream_count']/data['playlist_customers_no']  
    data['playlist_passion_score']=data['playlist_passion_score'].fillna(0)
    return data

def getPlaylistFeatures(data):
    data = playlist_avg_stream_counts(data)
    data = playlist_avg_number_of_users(data)
    data = playlist_avg_passion_score(data)
    return data

def getArtistPlaylistFeatures(data):
    # The above features are not yet averaged per artist. 
    # So it won't be valid once we drop the duplicate artist names
    data = playlist_avg_stream_counts(data)
    data = playlist_avg_number_of_users(data)
    data = playlist_avg_passion_score(data)
    ap_stream_count = data.groupby('artist_name')['playlist_stream_count'].mean().to_dict()
    ap_passion = data.groupby('artist_name')['playlist_passion_score'].mean().to_dict()
    ap_customers = data.groupby('artist_name')['playlist_customers_no'].mean().to_dict()
    
    data['ap_stream_count']=data['artist_name'].map(ap_stream_count)
    data['ap_stream_count']=data['ap_stream_count'].fillna(0)    
    data['ap_passion']=data['artist_name'].map(ap_passion)
    data['ap_passion']=data['ap_passion'].fillna(0)
    data['ap_customers']=data['artist_name'].map(ap_customers)
    data['ap_customers']=data['ap_customers'].fillna(0)
    
    data = data.drop(['playlist_customers_no','playlist_customers_no', 'playlist_passion_score'],axis=1)
    return data
```

```python
# Gender breakdown
def gender_break(datain):
    data = datain.drop_duplicates('customer_id')
    
    gender_dict = pd.crosstab(data['gender'], data['artist_name']).to_dict()
    break_dict = {}

    for key in gender_dict:
        sub = gender_dict[key]
        break_dict[key] = sub['male'] / (sub['female'] + sub['male'])
        
    data['gender_break'] = data['artist_name'].map(break_dict)
    return data

# Age breakdown
def age_percentages(datain):
    data = datain.drop_duplicates('customer_id').copy()
    data['birth_year']=data['birth_year'].dropna()
    data['age_bin'] = pd.cut((2017 - data.birth_year),np.arange(0,100,20)).to_frame()
    
    
    age_dict = pd.crosstab(data['age_bin'], data['artist_name']).to_dict()
    
    bin_keys = pd.cut((2017 - data.birth_year),np.arange(0,100,20)).unique()
    bin_keys = bin_keys.dropna()

    for bin_keym in bin_keys:
        percent_dict = {}

        for key in age_dict:
            
            summ = 0
            for bin_key in bin_keys:

                num = age_dict[key][bin_key]
                summ = summ + num
            
            percent_dict[key] =  age_dict[key][bin_keym]/summ
            new_bin_key = str(bin_keym.left) + str(bin_keym.right)
        data[new_bin_key] = data['artist_name'].map(percent_dict)  
    data = data.drop(['birth_year'], axis=1)
    return data
        
def age_percentages_10bins(datain):
    data = datain.drop_duplicates('customer_id').copy()
    data['birth_year']=data['birth_year'].dropna()
    data['age_bin'] = pd.cut((2017 - data.birth_year),np.arange(0,100,10)).to_frame()
    
    
    age_dict = pd.crosstab(data['age_bin'], data['artist_name']).to_dict()
    
    bin_keys = pd.cut((2017 - data.birth_year),np.arange(0,100,10)).unique()
    bin_keys = bin_keys.dropna()

    for bin_keym in bin_keys:
        percent_dict = {}

        for key in age_dict:
            
            summ = 0
            for bin_key in bin_keys:

                num = age_dict[key][bin_key]
                summ = summ + num
            
            percent_dict[key] =  age_dict[key][bin_keym]/summ
            new_bin_key = str(bin_keym.left) + str(bin_keym.right)
        data[new_bin_key] = data['artist_name'].map(percent_dict)  
    data = data.drop(['birth_year'], axis=1)
    return data
```

**Principle Component Analysis**

The data also contains a partial region code of the listener. We might want to consider including the regional breakdown of streams per artist as a feature of our model, to know if streams for certain regions are particularly influential on the future performance of an artist. 

However, we have over 400 unique regions and like playlists, including them all would lead to too many features and a large sparse matrix. One way in which to extract relevant ‘generalized’ features of each region would be to incorporate census and demographic data, from publicly available datasets. 

This is however beyond the scope of this courswork. Instead, a better way to summarize the impact of regional variation in streams is to use dimensionality reduction techniques. Here we will use Principle Component Analysis (PCA) to capture the regional variation in stream count.

PCA captures the majority of variation in the original feature set and represents it as a set of new orthogonal variables. Each ‘component’ of PCA is a linear combination of every feature, i.e. playlist in the dataset. Use **`scikit-learn`**’s PCA module (Pedregosa, et al., 2011) for generating PCA components.

The below section will find the PCA features for a given column input. We chose to do the analysis on region and playlist in order to find the top playlists/regions that explained the most variance in play-counts. Given the structure of the data, we must first turn the data into a matrix where each artist is a row, and each column is the playlist/region code. The values in the matrix are the play counts associated with that combination of arist and playlist/region. From there we created a graph that showed the amount of variance explained by the number of components and we were able to chose 10 regions and 20 playlists as with these numbers roughly 80% of variance was explained. Furthermore, we then extracted the PCA values and append them to our data set to be used as features in the model. 

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def create_matrix(df,row_name,col_name):
    '''Creates a matrix needed for PCA.
    Enter row_name, and col_name as strings'''
    dataframe = df[[row_name,col_name]]
    matrix = pd.pivot_table(dataframe,index = row_name,columns = col_name, aggfunc=len,fill_value = 0)
    return matrix

def standardize(matrix):
    '''Takes in a matrix and returns a standaridzed matrix.'''
    scaler = StandardScaler().fit(matrix)
    rescaled_matrix = scaler.transform(matrix)
    return rescaled_matrix

#PCA
def PCA_analysis(matrix,rescaled_matrix,n_components):
    '''Takes rescaled matrix and runs initial PCA.
    Returns final df with PCA cols'''
    pca = PCA(n_components = n_components)
    fit = pca.fit(rescaled_matrix)
    d2 = pca.transform(rescaled_matrix)
    d2_df = pd.DataFrame(d2)
    d2_df.index = pd.DataFrame(data=matrix).index
    col_headers = []
    for x in range(n_components):
        col_headers.append('PC'+str(x+1))
    d2_df.columns = col_headers
    #Plot 100 components
    pca_big = PCA(n_components = 100)
    fit_big = pca_big.fit(rescaled_matrix)
    var = np.cumsum(np.round(pca_big.explained_variance_ratio_,decimals=3)*100)
    plt.ylabel('% Variance Explained')
    plt.xlabel('# of Features')
    plt.title('PCA Analysis')
    plt.ylim(0,100.5)
    plt.style.context('seaborn-whitegrid')
    plt.plot(var)
    return d2_df

#Put it all together
def agg_PCA(df,row_name,col_name,n_components):
    m = create_matrix(df,row_name,col_name)
    rm = standardize(m)
    df_final = PCA_analysis(m,rm,n_components)
    
    return df_final

def PCA_merge(data,pca1,pca2):
    data = pd.merge(data, pca1, how='left', left_on='artist_name', left_index=False, right_index=True)
    data = pd.merge(data, pca2, how='left', left_on='artist_name', left_index=False, right_index=True)
    return data
```

Since we want to make sure have consistent threshold for dropping multicollinear columns, we wrote the `remove_high_corr` function below to automate the columns removal.

```python
def remove_high_corr(data, tresh):
    
    cor = data.corr()[(abs(data.corr()) >= tresh) & (data.corr() != 1)].dropna(how='all').dropna(axis=1,how='all')
    
    cor_list1 = []
    cor_list2 = []

    for id1 in cor.index.tolist():
        for id2 in cor.index.tolist():
            if (~np.isnan(cor[id1][id2])):
            
                cor_list1.append(id1)
                cor_list2.append(id2)
                
    full = data.corr().succ
    drop_list = []

    for i in range(0, len(cor_list1)):
        id1 = cor_list1[i]
        id2 = cor_list2[i]
    
        val1 = full[id1]
        val2 = full[id2]
    
        if (np.isnan(val1) | np.isnan(val2)):
            pass
    
        if((val1 > val2) & (val1 not in drop_list)):
            drop_list.append(id1)
        elif (val2 not in drop_list):
            drop_list.append(id2)
        
    drop_list = list(filter(lambda i: i != 'succ', drop_list))
    print('\ndropping collinear columns', drop_list)
    data = data.drop(drop_list,axis=1)
    print('final columns list: ', data.columns)
        
```

```python
# Class balance
from sklearn.utils import resample

# succ_rat: 0.0 - 1.0
# if 0.6, 60% of final data would be succ
def up_succ(data, succ_rat):
    
    data_succ = data[data.succ == 1]
    data_not_succ = data[data.succ == 0]
    
    data_upped = resample(data_succ,
                         replace = True,
                         n_samples = int(0.5*(len(data_not_succ) * succ_rat) / (1 - succ_rat)))
    data_upped_ns = resample(data_not_succ,
                         replace = True,
                         n_samples = int(0.5*(len(data_not_succ))))
    
    data_out = pd.concat([data_upped_ns, data_upped])
    print(data_out.groupby('succ')['succ'].count())
    return data_out

def dn_succ(data, succ_rat):
    
    data_succ = data[data.succ == 1]
    data_not_succ = data[data.succ == 0]
    data_downed = resample(data_not_succ,
                          replace = True,
                          n_samples = int((len(data_succ) * (1 - succ_rat))/succ_rat))
    
    data_out = pd.concat([data_succ, data_downed])

    return data_out
```
